```python
from pyspark.sql.functions import *
```

```python
df_emp=spark.read.table('employees')
df_leaves=spark.read.table('leaves')
df_payroll=spark.read.table('payroll')
```

```python
dataframes = [df_emp, df_leaves, df_payroll]
cleaned_dataframes = []
```

```python
for df in dataframes:
    for c in df.columns:
        if dict(df.dtypes)[c] == 'string':
            df = df.withColumn(c, trim(col(c)))
    cleaned_dataframes.append(df)
```

```python

df_emp, df_leaves, df_payroll = cleaned_dataframes
```

```python
df_emp = df_emp.withColumn('full_name', concat(col('first_name'), lit(' '), col('last_name')))

df_emp=df_emp.drop_duplicates(['id'])
df_leaves=df_leaves.drop_duplicates(['id'])
df_payroll=df_payroll.drop_duplicates(['id'])

df_emp=df_emp.withColumn('Loaded_at',current_date())
df_leaves=df_leaves.withColumn('Loaded_at',current_date())
df_payroll=df_payroll.withColumn('Loaded_at',current_date())

display(df_emp)
display(df_leaves)
display(df_payroll)

df_emp.write.mode("overwrite").format("delta").saveAsTable("Silver_Employee")
df_leaves.write.mode("overwrite").format("delta").saveAsTable("Silver_leaves")
df_payroll.write.mode("overwrite").format("delta").saveAsTable("Silver_payroll")
```
